```python
#сортировка пузырьком
list_time_bubble = []  #list with time correcponing list_int size of vector

for i in list_int:
  time_start = time.time()
  for _ in range(5):  #generat 5 times
    #data = [(lambda i: i)(i) for i in range(i)]
    a = []
    for m in range(i):  #generate random list
      a.append(random.randint(1, 100))
 
    for k in range(i-1):
      for l in range(i-k-1):
          if a[l] > a[l+1]:
              a[l], a[l+1] = a[l+1], a[l]
     
  time_stop = time.time()
  list_time_bubble.append(time_stop - time_start)
list_time_bubble

 
```

```python
df['time of bubble sort'] = list_time_bubble
sns.scatterplot(data=df,x='size of vector',y='time of bubble sort')
```

```python
def quicksort(nums):
   if len(nums) <= 1:
       return nums
   else:
       q = random.choice(nums)
       s_nums = []
       m_nums = []
       e_nums = []
       for n in nums:
           if n < q:
               s_nums.append(n)
           elif n > q:
               m_nums.append(n)
           else:
               e_nums.append(n)
       return quicksort(s_nums) + e_nums + quicksort(m_nums)
```

```python
#сортировка пузырьком
list_time_quick = []  #list with time correcponing list_int size of vector

for i in list_int:
  time_start = time.time()
  for _ in range(5):  #generat 5 times
    #data = [(lambda i: i)(i) for i in range(i)]
    a = []
    for m in range(i):  #generate random list
      a.append(random.randint(1, 100))
 
    quicksort(a)
     
  time_stop = time.time()
  list_time_quick.append(time_stop - time_start)
list_time_quick
```

```python
df['time of quicksort'] = list_time_quick
sns.scatterplot(data=df,x='size of vector',y='time of quicksort')
```

```python
#сортировка timsort
list_time_tim = []  #list with time correcponing list_int size of vector

for i in list_int:
  time_start = time.time()
  for _ in range(5):  #generat 5 times
    #data = [(lambda i: i)(i) for i in range(i)]
    a = []
    for m in range(i):  #generate random list
      a.append(random.randint(1, 100))
 
    a.sort()  #standart realisation the quickest sorting
     
  time_stop = time.time()
  list_time_tim.append(time_stop - time_start)
list_time_tim
```

```python
df['time of timsort'] = list_time_tim
sns.scatterplot(data=df,x='size of vector',y='time of timsort')
```
